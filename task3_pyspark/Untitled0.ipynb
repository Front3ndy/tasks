```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.master('local').appName('task_spark').getOrCreate()
```

```python
category_data = [
    (1, 'fruits'),
    (2, 'vegetables'),
    (3, 'milk_products')
]
category_schema = ['id', 'category_name']
category = spark.createDataFrame(category_data, category_schema)
```

```python
category.show()
```

```python
product_data = [
    (1, 'banana', 1, 10),
    (2, 'tomato', 2, 5),
    (3, 'cheese', 3, 25),
    (4, 'bread', 5, 15)
]
product_schema = ['id', 'product_name', 'category_id', 'price']
product = spark.createDataFrame(product_data, product_schema)
```

```python
product.show()
```

```python
product.join(category, 'id', 'left').show()
```
